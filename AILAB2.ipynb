```python
# create two sets
set1 = {1, 2, 3, 4, 5}
set2 = {4, 5, 6, 7, 8}

print("Union:", set1 | set2)          # union
print("Intersection:", set1 & set2)   # intersection
print("Difference:", set1 - set2)     # difference
```

```python
# Create a dictionary with 5 keys
my_dict = {
    "name": "Ali",
    "age": 25,
    "city": "Lahore",
    "job": "Engineer",
    "hobby": "Reading"
}

print("Original:", my_dict)

# Update the dictionary
my_dict["age"] = 26  # change age
my_dict["city"] = "Islamabad"  # change city

print("Updated:", my_dict)
```

```python
# create a tuple with 5 fruit names
fruits = ("apple", "banana", "mango", "orange", "grape")

# print the second element
print("Second fruit:", fruits[1])

# print the last element
print("Last fruit:", fruits[-1])
```

```python
# list of 10 integers
numbers = [5, 2, 9, 1, 7, 3, 8, 6, 4, 10]

print("Original List:", numbers)
print("sum:",sum(numbers))
print("max:",max(numbers))
print("min:",min(numbers))

# add one number
numbers.append(11)
print("After Adding:", numbers)
# delete one number
numbers.remove(7)

print("After Deletion:", numbers)


# ascending order
numbers.sort()
print("Ascending Order:", numbers)
numbers.sort(reverse=True)
print("Descending:", numbers)
while False:
    print("This runs forever")

```
